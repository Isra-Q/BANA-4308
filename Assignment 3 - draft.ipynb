```python
# This addresses the first issue of getting the hyper links.
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

#url = input('Enter - ')
url = 'https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc'
html = requests.get(url, headers = headers)
soup = BeautifulSoup(html.content, "html.parser")
#using attrs = {'':''} allows us to use multiple attributes when webscrapping.
url_list = soup.find_all('a', attrs = {'class' : 'title'})
for i in url_list:
    print('https://www.metacritic.com' + i.get('href'))
```

```python
#This library helps use the time attributes
import time
#start a counter to know if you run into an error
counter = 1

# goes into each url to get the info we are looking for.

for i in url_list:
    print("Movie No: ", counter)
    counter +=1
    movie_url = ('https://www.metacritic.com' + i.get('href'))
    html = requests.get(movie_url, headers = headers)
    #soup will be the page of each movie one by one.
    soup = BeautifulSoup(html.content, "html.parser")
    
#Can use find in this step if we know that we will only find one director.
    
    #Director Divider = try and except statement in case there is no director.
    director_divider = soup.find('div', attrs = {'class': 'director'})
    #print(director_divider)
    
    #Looking for 'a' tag inside director divider list (1st element)
    director = director_divider.find_all('a')
    for i in director:
        print("Director(s): ",i.get_text())
    #time.sleep(0.5) delays the time between the loop itterations
    geres_divider = soup.find('div', attrs = {'class':'genres'})
    genres = geres_divider.find_all('span')
    lst = []
    for i in genres:
        lst.append(i.get_text())
    print("Genre: ", lst[2:])
    score_tag = soup.find('a', attrs = {'class':'metascore_anchor'})
    score = score_tag.get_text()
    print("Meta Score: ",score)
    time.sleep(1)
```

```python
page_count = 0

for i in range(0,5):
    url2 = 'https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=' + str(page_count)
    page_count += 1
    html2 = requests.get(url2, headers = headers)
    soup2 = BeautifulSoup(html2.content, "html.parser")
    #using attrs = {'':''} allows us to use multiple attributes when webscrapping.
    url_list2 = soup2.find_all('a', attrs = {'class' : 'title'})
    
    for i in url_list2:
        all_movies = 'https://www.metacritic.com' + i.get('href')
        print(f'Movies:  {all_movies}')
```
